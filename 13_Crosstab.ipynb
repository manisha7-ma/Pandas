```python
'''
A **contingency table**, also known as a cross-tabulation or crosstab, is a type of table in a matrix format that displays the multivariate frequency distribution of the variables. It is a simple yet powerful tool used in statistics to analyze the relationship between two or more categorical variables.

Think of it as a way to summarize and visualize how the values of one categorical variable are distributed across the values of another categorical variable.

**Structure of a Contingency Table:**

A typical contingency table has:

*   **Rows:** Represent the categories of one variable.
*   **Columns:** Represent the categories of another variable.
*   **Cells:** Each cell in the table contains the count or frequency of observations that fall into the intersection of a specific row category and a specific column category.
*   **Marginal Totals:** The sums of the counts in each row and each column are called marginal totals. These represent the total frequency for each category of the individual variables.

**Example to Illustrate:**

Let's consider a simple example involving two categorical variables: **Gender** (Male, Female) and **Smoking Status** (Smoker, Non-smoker).

A contingency table for this could look like this:

| Smoking Status | Male | Female | Total |
|---|---|---|---|
| Smoker | 150 | 100 | 250 |
| Non-smoker | 350 | 400 | 750 |
| **Total** | 500 | 500 | 1000 |

*   **Rows:** Smoker, Non-smoker
*   **Columns:** Male, Female
*   **Cells:**
    *   150: Number of Male Smokers
    *   100: Number of Female Smokers
    *   350: Number of Male Non-smokers
    *   400: Number of Female Non-smokers
*   **Marginal Totals:**
    *   250: Total Smokers
    *   750: Total Non-smokers
    *   500: Total Males
    *   500: Total Females
    *   1000: Grand Total (Total number of people surveyed)

**Purpose and Usage of Contingency Tables:**

Contingency tables are primarily used to:

1.  **Summarize and Visualize Data:** They provide a clear and organized way to see the distribution of observations across different categories of two or more variables. This makes it easier to understand the data at a glance.

2.  **Identify Relationships Between Variables:** By looking at the frequencies in the cells, you can start to see if there's a relationship or association between the variables. For instance, in the example above, you might observe a higher proportion of smokers among males compared to females (comparing 150/500 vs 100/500 or 150/250 vs 100/250).

3.  **Perform Statistical Tests:** Contingency tables are the foundation for various statistical tests used to determine if the observed relationship between the variables is statistically significant (meaning it's unlikely due to random chance). Common tests include:
    *   **Chi-Squared Test for Independence:** This test checks if the two categorical variables are independent (no association) or dependent (associated).
    *   **Fisher's Exact Test:** Used for small sample sizes, similar to the chi-squared test.

**Your Example (Country, Age, Sex, Handedness):**

To apply the concept to your example (Country, Age, Sex, Handedness), a contingency table could involve three or four variables. For example, you could create a 3-way contingency table to examine the relationship between **Country**, **Sex**, and **Handedness**.

*   **Variables:** Country (e.g., USA, UK, India), Sex (Male, Female), Handedness (Left, Right)
*   You would have separate tables or multi-dimensional arrays to represent this. A simplified 2-way table could look at the relationship between **Sex** and **Handedness** *within* a specific **Country**.

For instance, a table for the USA:

| Handedness | Male (USA) | Female (USA) | Total (USA) |
|---|---|---|---|
| Left | 100 | 80 | 180 |
| Right | 900 | 920 | 1820 |
| **Total (USA)** | 1000 | 1000 | 2000 |

'''
```

```python
# prompt: syntax of crosstab in python with all parameters

# The basic syntax is:
# pandas.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, dropna=True, normalize=False)

# Parameters:
# index: array-like, or list of array-likes
#        Arrays or lists of arrays to group by on the rows.
# columns: array-like, or list of array-likes
#        Arrays or lists of arrays to group by on the columns.
# values: array-like, optional
#        Array of values to aggregate according to the factors. Requires `aggfunc`.
# rownames: sequence, optional
#        If passed, must match number of row arrays passed.
# colnames: sequence, optional
#        If passed, must match number of column arrays passed.
# aggfunc: function, optional
#        If specified, requires `values`. If `values` is not specified, `aggfunc` will be ignored.
# margins: bool, default False
#        Add row/column margins (subtotals).
# dropna: bool, default True
#        Do not include columns whose entries are all NaN.
# normalize: bool, {'all', 'index', 'columns'}, default False
#        Normalize by dividing all values by the sum of values.
#        - If False, do not normalize.
#        - If 'all', normalize by the sum of all values.
#        - If 'index', normalize by the sum of each row.
#        - If 'columns', normalize by the sum of each column.

```

Basically Pass in the column or the variable who's frequency you want to see along with the column parameter on top of which it is being visualised

```python
pd.crosstab(df.Sex,df.Handedness ,margins=True)
```

```python
#If I want multiple variable in column level
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
```

```python

pd.crosstab([df.Handedness,df.Sex],df.Nationality)
```

```python
#Having it in percentage
# normalize: bool, {'all', 'index', 'columns'}, default False
#        Normalize by dividing all values by the sum of values.
#        - If False, do not normalize.
#        - If 'all', normalize by the sum of all values.
#        - If 'index', normalize by the sum of each row.
#        - If 'columns', normalize by the sum of each column.
pd.crosstab(df.Sex,df.Handedness,normalize='index')
```

```python
#knowing the average age of individual per sex
# values: array-like, optional
#        Array of values to aggregate according to the factors. Requires `aggfunc`.

import numpy as np
pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
```
