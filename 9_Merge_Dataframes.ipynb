```python
import pandas as pd
```

```python
#merge in pandas
df1=pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[21,14,35]
})
df2=pd.DataFrame({
    "city":["chicago","newyork","orlando"],
    "humidity":[65,68,75]
})
df=pd.merge(df1,df2,on="city")
df
```

```python
#Now seeing the case when the number of rows is different in both the Dataframes
df1=pd.DataFrame({
    "city":["newyork","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38]
})
df2=pd.DataFrame({
    "city":["chicago","newyork","san francisco"],
    "humidity":[65,68,71]
})
df=pd.merge(df1,df2,on="city")
df
```

```python
#by default it takes the intersection of the both dataframes object
#outer join
df=pd.merge(df1,df2,on="city",how="outer")
df
```

```python
#left join
df=pd.merge(df1,df2,on="city",how="left")
```

```python
df
```

```python
#right Join
df=pd.merge(df1,df2,on="city",how="right")
df
```

```python
#getting the Indicator as to which part it is coming from
df=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df
```

```python
#suffices
#If we have common columns
df1=pd.DataFrame({
    "city":["newyork","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38],
    "humidity":[65,68,71,75]
})
df2=pd.DataFrame({
    "city":["chicago","newyork","san francisco"],
    "temperature":[21,14,35],
    "humidity":[65,68,71]
})
#Using Suffixes parameter we can specify the column suffix
#to identify where it is coming from
df=pd.merge(df1,df2,on="city",how="outer",suffixes=("_first","_second"))
df
```
