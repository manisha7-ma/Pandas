```python
#DataFrame Shaping Technique
#Stack   UnStack

import pandas as pd
df=pd.read_excel("/content/sample_data/stocks.xlsx",header=[0,1])
df
```

```python
#Needing one of the column level and transform it into a row
#By deafult when we are running the stck function it takes the
#innermost column in consideration
df.stack()
```

```python
#doing stacking on the level 0
```

```python
df.stack(level=0)
```

```python
#using dropna =true to get rid
```
