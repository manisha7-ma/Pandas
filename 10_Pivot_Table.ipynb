```python
#Pivot and Pivot Table
#Pivot allows us to reshape or transform data
import pandas as pd
df=pd.read_csv('/content/sample_data/weather.csv')
df.head()
```

```python
#transforming it
#index -- rows which will be used to transform
#columns -- column that  will be used for categorisation
#values --the columns under which it will be categorised
df.pivot(index='date',columns='city')
```

```python
#only showing the humididty column
df.pivot(index='date',columns='city',values='humidity')
```

```python
#Pivot Table
'''
Pivot table allows us to summarize and aggregate data inside a dataframe
'''
df=pd.read_csv('/content/sample_data/weather2.csv')
df
```

```python
#In the above dataframe we have same day temerature and
#humidity of the same city
#we can aggreagte this using the pivot table concept
df=pd.pivot_table(df,index='city',columns='date')
df
```

```python
#It by default took the average of the values of that column for that date
#pivot table we supply aggFunc
```

```python
#default aggfunc = mean
#It can also be aggFunc=sum etc
#Margin =True ,Margin=False
#Margin provides the aggregate per row and column
```

```python
#Grouper
df=pd.read_csv('/content/sample_data/weather3.csv')
df
```

```python
#Let say I want aggregate temperature per month
#I want aggreagte based on Date
#let say aggreagte of temp per month
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
