```python
#Concatenation of DataFrames
import pandas as pd
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
#concatenating the two dataframes
df=pd.concat([india_weather,us_weather])
df
```

```python
#If we see the index being used is just repeated
#Let us have them all index in sequence
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
#Trying to get the Dataframes per country
#passing Keys : Associate it with each Dataframe
df=pd.concat([india_weather,us_weather],keys=["india","us"])
```

```python
df
```

```python
#getting the values from the merged DataFrame based on key passed
df.loc["us"]
```

```python
#vertical Concatenation
temperature_df=pd.DataFrame({
    "city":["mumbai","kanpur","patna"],
    "temperature":[32,45,30]
},index=[0,1,2])
windspeed_df=pd.DataFrame({
    "city":["kanpur","patna"],
    "windspeed":[7,12]
},index=[1,2])
```

```python
#Axis parameter by default is 0 for horizontal concatenation
#It has to be passed as 1 for vertical concat
df=pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
df
```

```python
#joining dataframe object with a series
s=pd.Series(["Humid","Dry"],name="Event",index=[0,2])
```

```python
#concatenating dataframe and series
df=pd.concat([temperature_df,s],axis=1)
```
