```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/sample_data/weather_data.csv")
```

```python
df
```

```python
#Task is to replace these weird value -9999 with something more meaningful
```

```python
#Replacing with Nan
#It is done using the replace function
#The first parameter in the replace function is the value that has to be replaced
#it could be a single value or a list
#The second parameter is the Value with which it has to be replaced with
df.replace(-99999,np.nan)
```

```python
#Replacing values based on specific columns
newdf=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.nan)
```

```python
newdf
```

```python
newdf=newdf.replace({
    'event':np.nan
},'No Event'
                    )
```

```python
newdf
```

```python
#replacing values based on values
df3=df.replace({
    -99999:np.nan,
    'No Event':'Sunny'
})
```

```python
df3
```

```python
#Using Regex to replace the values of the units if any present after the values
df4=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
```

```python
df4
```

```python
#replacing list of values with another list of values
af=pd.DataFrame({
    'score':['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','maya','parthiv','tom','julian','erica']
})
```

```python
af
```

```python
#replacing the entire list with list of values
af.replace(['poor','average','exceptional','good'],[1,2,3,4])
```
